```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Specify the subset of columns
columns_to_fill = ['G', 'PA', 'HR', 'R', 'RBI', 'SB', 'GDP', 'CS', 'H',
                   'G_2', 'PA_2', 'HR_2', 'R_2', 'RBI_2', 'SB_2', 'GDP_2', 'CS_2', 'H_2',
                   'G_3', 'PA_3', 'HR_3', 'R_3', 'RBI_3', 'SB_3', 'GDP_3', 'CS_3', 'H_3']

# Calculate the median for the specified columns
median_values = df[columns_to_fill].median()

# Fill missing values with the median for the specified columns
df[columns_to_fill] = df[columns_to_fill].fillna(median_values)

# Check for missing values after filling with the median
missing_values_after_filling = df[columns_to_fill].isnull().sum()

# Display the number of missing values after filling
print("Missing Values After Filling with Median:")
print(missing_values_after_filling)

```

```python
import numpy as np

# Specify the float variables to be converted to integers
float_to_int_columns = ['G', 'PA', 'HR', 'R', 'RBI', 'SB', 'GDP', 'CS', 'H',
                        'G_2', 'PA_2', 'HR_2', 'R_2', 'RBI_2', 'SB_2', 'GDP_2', 'CS_2', 'H_2',
                        'G_3', 'PA_3', 'HR_3', 'R_3', 'RBI_3', 'SB_3', 'GDP_3', 'CS_3', 'H_3']

# Replace non-finite values with 0
df[float_to_int_columns] = df[float_to_int_columns].fillna(0).astype(int)

# Check the data types after conversion
df.dtypes
```
